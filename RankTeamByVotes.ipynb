## 1524. Number of Sub-arrays With Odd Sum

```python
class Solution:
    def createSubarray(self, arr, curr_arr, ans, i):
        if i==len(arr):
            ans.append(curr_arr)
            return
        
        self.createSubarray(arr, curr_arr + [arr[i]], ans, i+1)
        self.createSubarray(arr, curr_arr, ans, i+1)
        
        return ans
    
    def numOfSubarrays(self, arr):
        odd_sum = 0
        subArrays = self.createSubarray(arr, [], [], 0)
        subArrays = [list(item) for item in set(tuple(row) for row in subArrays)]
        print(subArrays)
        for subArray in subArrays:
                if sum(subArray)%2 != 0:
                    odd_sum+=1
        return odd_sum
        
s = Solution()
s.numOfSubarrays([1,2,3,4,5,6,7])
```

## NUMBER LOWER THAN THE CURRENT NUMBER

```python
class Solution:
    def positionOfNumInArr(self, arr, num):
        low = 0
        high = len(arr)-1
        while(low<=high):
            middle = (low + high)//2
            
            if (arr[middle] == num and middle-1>=0 and arr[middle-1]!=num) or (middle==0 and arr[middle]==num):
                return middle
            
            elif arr[middle]>=num:
                high = middle-1
            else:
                low = middle+1
        return -1
    
    def smallerNumbersThanCurrent(self, nums):
        
        sorted_nums = sorted(nums)
        
        ans = []
        
        for num in nums:
            pos = self.positionOfNumInArr(sorted_nums, num)
            ans.append(max(0, pos))
        
        return ans
```

## RAIN WATER TRAPPING

```python
class Solution:

    def trap(self, height):
        rMax = []
        lMax = []
        for i in range(len(height)):
            _max = height[i]
            for j in range(i+1, len(height)):
                _max = max(_max, height[j])
                    
            rMax.append(_max)
            
        for i in range(len(height)-1, -1, -1):
            _max = height[i]
            for j in range(i-1, -1,-1):
                _max = max(_max, height[j])
            lMax.append(_max)
        lMax.reverse()
            
        print(rMax)
        print(lMax)
       
    
        
```
